```python
# Dependencies and Setup
import pandas as pd

# File to Load (Remember to Change These)
school_data_to_load = "Resources/schools_complete.csv"
student_data_to_load = "Resources/students_complete.csv"

# Read School and Student Data File and store into Pandas DataFrames
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)

# Combine the data into a single dataset.  
school_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
school_data_complete
```

## District Summary

* Calculate the total number of schools

* Calculate the total number of students

* Calculate the total budget

* Calculate the average math score 

* Calculate the average reading score

* Calculate the percentage of students with a passing math score (70 or greater)

* Calculate the percentage of students with a passing reading score (70 or greater)

* Calculate the percentage of students who passed math **and** reading (% Overall Passing)

* Create a dataframe to hold the above results

* Optional: give the displayed data cleaner formatting

```python
sum_schools = school_data_complete["school_name"].nunique()
#sum_schools

sum_students = school_data_complete["student_name"].count()
#sum_students

budget= school_data_complete["budget"].unique()
total_budget = budget.sum()
total_budget

avg_math_score = school_data_complete["math_score"].mean()
#avg_math_score

avg_reading_score = school_data_complete["reading_score"].mean()
#avg_reading_score

passing_math = school_data_complete.loc[school_data_complete["math_score"] >= 70]["math_score"].count()
perc_passing_math = passing_math/sum_students*100
#perc_passing_math

passing_reading = school_data_complete.loc[school_data_complete["reading_score"] >= 70]["reading_score"].count()
perc_passing_reading = passing_reading/sum_students*100
#perc_passing_reading

perc_student_passing = school_data_complete[(school_data_complete["math_score"] >= 70) & (school_data_complete["reading_score"] >= 70)].count()["student_name"]/sum_students*100
#perc_student_passing

district_df = pd.DataFrame({"Total Schools": [sum_schools],
    "Total Students": [sum_students],
    "Total Budget": [total_budget],
    "Average Reading Score": [avg_math_score],
    "Average Math Score": [avg_reading_score],
    "% Passing Reading":[perc_passing_math],
    "% Passing Math": [perc_passing_reading],
    "Overall Passing Rate": [perc_student_passing]})
#district_df

#Get it on with the formatting 
district_df["Total Schools"]= district_df["Total Schools"].map("{:.2f}".format)
district_df["Total Students"]= district_df["Total Students"].map("{:.0f}".format) 
district_df["Total Budget"]= district_df["Total Budget"].map("${:.0f}".format) 
district_df["Average Reading Score"]= district_df["Average Reading Score"].map("{:.1f}".format) 
district_df["Average Math Score"]= district_df["Average Math Score"].map("{:.1f}".format) 
district_df["% Passing Reading"] = district_df["% Passing Reading"].map("{:.1f}%".format)
district_df["% Passing Math"] = district_df["% Passing Math"].map("{:.1f}%".format)
district_df["Overall Passing Rate"] = district_df["Overall Passing Rate"].map("{:.1f}%".format)
district_df
```

## Scores by School Spending

* Create a table that breaks down school performances based on average Spending Ranges (Per Student). Use 4 reasonable bins to group school spending. Include in the table each of the following:
  * Average Math Score
  * Average Reading Score
  * % Passing Math
  * % Passing Reading
  * Overall Passing Rate (Average of the above two)

```python
spend_bins = [0, 584.999, 614.999, 644.999, 999999]
bin_names = ["< $585", "$585 - 614", "$615 - 644", "> $644"]

school_data_complete["spend_bins"] = pd.cut(school_data_complete["size"], spend_bins, labels = bin_names)

spend_info = school_data_complete.groupby("spend_bins")

##########################################################

avg_math = spend_info["math_score"].mean()
avg_read = spend_info["reading_score"].mean()
pass_math = school_data_complete[school_data_complete["math_score"] >= 70].groupby("spend_bins")["Student ID"].count()/spend_info["Student ID"].count()*100
pass_read = school_data_complete[school_data_complete["reading_score"] >= 70].groupby("spend_bins")["Student ID"].count()/spend_info["Student ID"].count()*100
overall = school_data_complete[(school_data_complete["reading_score"] >= 70) & (school_data_complete["math_score"] >= 70)].groupby("spend_bins")["Student ID"].count()/spend_info["Student ID"].count()*100

########################################################## 
    
spend_scores = pd.DataFrame({
    "Average Math Score": avg_math,
    "Average Reading Score": avg_read,
    "% Passing Math": pass_math,
    "% Passing Reading": pass_read,
    "Overall Passing Rate": overall})

##
    
spend_scores["Average Math Score"] = spend_scores["Average Math Score"].map("{:.1f}%".format)
spend_scores["Average Reading Score"] = spend_scores["Average Reading Score"].map("{:.1f}%".format)
spend_scores["% Passing Math"] = spend_scores["% Passing Math"].map("{:.1f}%".format)
spend_scores["% Passing Reading"] = spend_scores["% Passing Reading"].map("{:.1f}%".format)             
spend_scores["Overall Passing Rate"] = spend_scores["Overall Passing Rate"].map("{:.1f}%".format)              

spend_scores
```
